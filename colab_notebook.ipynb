# 🧠 Universal Sentence Encoder - TFLite Converter

**Adım 1:** Sol taraftaki 📁 dosya ikonuna tıkla
**Adım 2:** `universal-sentence-encoder-tensorflow2-multilingual-v2.tar.gz` dosyasını yükle
**Adım 3:** Aşağıdaki hücreleri sırayla çalıştır (▶️ butonuna bas)

````
# HÜCRE 1: TensorFlow Kur
!pip install -q tensorflow==2.14.0
print("✅ TensorFlow kuruldu!")
````

````
# HÜCRE 2: Modeli Çıkar
import tarfile
import os

tar = tarfile.open('universal-sentence-encoder-tensorflow2-multilingual-v2.tar.gz')
tar.extractall()
tar.close()

print("✅ Model çıkarıldı!")
print("Dosyalar:", os.listdir('.'))
````

````
# HÜCRE 3: TFLite'a Dönüştür
import tensorflow as tf

print("🔄 Dönüştürülüyor...")

converter = tf.lite.TFLiteConverter.from_saved_model('saved_model')
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()

# Kaydet
with open('use_multilingual_v2.tflite', 'wb') as f:
    f.write(tflite_model)

size_mb = len(tflite_model) / (1024 * 1024)
print(f"✅ Başarılı!")
print(f"📁 Çıktı: use_multilingual_v2.tflite")
print(f"📊 Boyut: {size_mb:.1f} MB")
````

````
# HÜCRE 4: İndir
from google.colab import files

print("📥 İndirme başlatılıyor...")
files.download('use_multilingual_v2.tflite')
print("✅ İndirme tamamlandı!")
````
